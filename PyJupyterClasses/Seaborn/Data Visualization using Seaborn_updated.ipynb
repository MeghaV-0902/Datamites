# Data Visulaization using Seaborn.

Seaborn is a library that uses Matplotlib underneath to plot graphs.
- Graphs that show the **relationship** between **two variables**-**Bivariate Graphs**
- Graphs that represent the **distribution** of a single variable.**Univariate Graphs**

```python
# importing the required libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# lineplot()

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=df,errorbar=None) #Confidence interval errorbar=('ci', 70)
plt.show()
```

# Scatter PLot

```python
## Scatter plot--distribution and relationships.
sns.scatterplot(x='sepal_length',y='sepal_width',data=df)
plt.title("Scatter plot")
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=df, hue='species')
plt.title("Scatter plot")
plt.show()
```

- Species Separation
Youâ€™ll likely see that Setosa clusters in a separate region (shorter and wider sepals).
Versicolor and Virginica may overlap more, but show different trends.
- correlation between sepal_length and sepal_width for each species.
A downward or upward slope suggests a trend.
Scatter spread shows variability within each group.
- Outliers or Anomalies
Points far away from the main cluster might be outliers or unusual observations.

# Bar PLot

```python
## Bar plot (categorical estimate plot)
##A barplot is basically used to aggregate the categorical data according to some methods and by default its the mean.
##It can also be understood as a visualization of the group by action.
#To use this plot we choose a categorical column for
##the x axis and a numerical column for the y axis and we see that it
# creates a plot taking a mean per categorical column.

sns.barplot(y='sepal_length', data=df,estimator=np.mean,errorbar=None)
plt.show()
```

```python
## Count Plot  value_count()
##A countplot basically counts the categories and returns a
# count of their occurrences.
sns.countplot(x='species', data=df)
plt.show()
```

# Box Plot

```python
## Box plot
##A boxplot is sometimes known as the box and whisker plot.It shows the
#distribution of the quantitative data
## that represents the comparisons between variables. boxplot shows the
#quartiles of the dataset while the whiskers extend to
## show the rest of the distribution i.e. the dots indicating the
#presence  of outliers.
sns.boxplot(x='sepal_width',data=df)
plt.show()
```

```python
df.describe()
```

A box plot (also called a whisker plot) summarizes the distribution of a numerical variable:

Box: 25th percentile (Q1) to 75th percentile (Q3) â†’ the interquartile range (IQR).

Line inside box: Median (50th percentile).

Whiskers: Extend to ~1.5 Ã— IQR beyond Q1 and Q3.

Dots outside whiskers: Outliers.

```python
sns.boxplot(x='species', y='sepal_width', data=df)
plt.show()
# Checking outliers in each category with respect to sepal width.
```

## Violinplot

```python
## Violinplot
## It is similar to the boxplot except that it provides a higher, more advanced visualization and uses the kernel
##density estimation to give a better description about the data distribution.

sns.violinplot(x='species', y='sepal_width', data=df)
plt.show()
```

## Stripplot

```python
##Stripplot
## It basically creates a scatter plot based on the category.
# It is created using the stripplot()
sns.stripplot(x='species', y='petal_width', data=df)
plt.show()
```

## 1. Raw Data Visibility
Unlike box or violin plots, this plot shows **every individual data point**.

Great for spotting:
- **Outliers**
- **Gaps or clusters**
- **Duplicates or tightly packed values**

---

## 2. Species-Level Differences
- **Setosa**: Most values clustered tightly around **0.2 cm**, showing very little variation.
- **Versicolor**: Ranges between **1.0 â€“ 1.8 cm**, with moderate spread.
- **Virginica**: Widest spread, going up to **2.5 cm**, indicating greater variation.

## Histogram

```python
## Histogram
## A histogram is basically used to represent data provided in a form of some groups.
## It is accurate method for the graphical representation of numerical data
## distribution
sns.histplot(x='petal_width', data=df,kde=True) # The data is divided into 5 equal-width intervals


plt.show()
```

```python
# categorical plot
sns.catplot(x='sepal_length', data=df, kind="bar")
plt.show()          # Options: "strip", "swarm", "box", "violin", "bar", "count", "point"
```

| plot     | Description                      | Use Case Example                  |
| -------- | -------------------------------- | --------------------------------- |
| `strip`  | Scatter of all observations      | Distribution of scores            |
| `swarm`  | Like `strip`, but avoids overlap | Compact view of individual points |
| `box`    | Boxplot with quartiles           | Visualizing spread and outliers   |
| `violin` | Combines KDE + boxplot           | Distribution + density            |
| `bar`    | Shows mean + CI                  | Comparing group averages          |
| `count`  | Barplot of count of observations | Frequency of categories           |
| `point`  | Shows means + CI with line       | Trends over time or ordered vars  |


Correlation studies and measures the direction and extent of relationship among variables
correlation summarize the relationship between two variables in a single number called the correlation coefficient. The correlation coefficient is usually represented using the symbol r, and it ranges from -1 to +1.

A correlation coefficient quite close to 0, but either positive or negative, implies little or no relationship between the two variables. A correlation coefficient close to plus 1 means a positive relationship between the two variables, with increases in one of the variables being associated with increases in the other variable.

A correlation coefficient close to -1 indicates a negative relationship between two variables, with an increase in one of the variables being associated with a decrease in the other variable.

```python
sns.heatmap(df.drop("species",axis=1).corr(),annot=True)
```

```python
##Heat map
## Heatmap is defined as a graphical representation of data using colors to visualize the value of the matrix.
## In this, to represent more common values or higher activities brighter colors basically reddish colors are used and
## to represent less common or activity values, darker colors are preferred

# correlation between the different parameters
sns.heatmap(df.corr(numeric_only=True), annot=True)#, cmap='coolwarm', fmt=".2f")

#Very strong linear relationship â€” as petal length increases, petal width also increases.

```
