### Map function
map(function_name, list or tuple)

```python
list_var=[1,3,4,5]
def squares(x):
    return x*x

result = map(squares, list_var) #Iterates over every element in the list or tuple without mentioning a loop
print(list(result))
```

```python
list_var=[1,3,4,5]
squaresL = lambda x : x*x

result = map(squaresL, list_var) #Iterates over every element in the list or tuple without mentioning a loop
print(list(result))
```

```python
list1=[1,2,3,4]
list2=[6,7,8,9]

result=map(lambda x,y : x+y, list1, list2)
print(list(result))
```
