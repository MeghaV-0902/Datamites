```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
d1=["SOP","LOR ","Research","GPA","Chance of Admit"]
data1= sc.fit_transform(data.drop(d1,axis=1))
```

```python
con_data= data[["SOP","LOR ","Research","GPA","Chance of Admit"]]
```

```python
con_data
```

```python
data.columns
```

```python
data2=pd.DataFrame(data1, columns=['GRE Score', 'TOEFL Score','University Rating'])
```

```python
data2
```

```python
final_df= pd.concat([data2,con_data],axis=1)
```

```python
final_df
```

```python
sns.heatmap(final_df.drop(["Chance of Admit"], axis=1).corr(), annot= True)
```

```python
data.columns
```

```python
X=final_df[['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'GPA', 'Research']]
y= final_df['Chance of Admit']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.2,random_state=23)
```

```python
from sklearn.linear_model import LinearRegression

lr= LinearRegression()
lr.fit(X_train,y_train)

y_pred=lr.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2score= r2_score(y_test,y_pred)
```
