# REGEX 

## What is Regex (Regular Expressions)?
       Regex (short for Regular Expressions) is like a search pattern used to match text.

     1.It‚Äôs basically a sequence of characters that defines a rule for finding strings inside text.

     2.Think of it like a CTRL+F on steroids ‚Üí much more powerful than normal search.

## Why Do We Need Regex?
-- In Machine Learning, Regex is not a model, but a preprocessing / feature engineering tool.We use it mostly in text-based ML tasks (NLP).

We need regex when we want to:

  1. Search ‚Üí Find a pattern in text

  2. Validate ‚Üí Check if input follows rules

  3. Extract ‚Üí Pull out useful pieces of text

  4. Replace/Clean ‚Üí Modify text based on patterns



```python
import pandas as pd

# Load CSV
df = pd.read_csv("reviews.csv")
df.head()
```

```python
import re
```

```python
df[df['Review'].str.contains(r"^love", flags=re.IGNORECASE, regex=True)]
```

```python
# Emails starting with ‚Äúbob‚Äù
df[df['Email'].str.contains(r'^bob', regex=True, na=False)]
```

```python
df[df['Phone'].astype(str).str.contains(r'^9', regex=True, na=False)]
```

```python
df[df['Name'].str.contains(r'^[AEIOUaeiou]', regex=True, na=False)]
```

```python
# $-Ends with
df[df['Email'].str.contains(r'example.com$', regex=True, na=False)]
```

```python
df[df['Phone'].astype(str).str.contains(r'5$', regex=True, na=False)]
```

```python
# Replace the values
df['Email'] = df['Email'].str.replace(r'[@.]', '_', regex=True)
```
