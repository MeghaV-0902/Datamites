# FEATURE ENGINEERING

1. Feature Engineering is the process of transforming raw data into meaningful inputs (features) that improve the performance of a Machine Learning model.
2. In other words, it’s about cleaning, modifying, and creating new features so that models can learn patterns more effectively.

## WHY IT IS IMPORTANT ?
1. Raw data often contains missing values, noise, outliers, or irrelevant information.
2. Machine Learning algorithms perform better when data is in the right scale, format, and representation.
3. The quality of features directly affects the accuracy and robustness of the model.

### HANDLING NULL VALUES
Why do missing values occur?
1. Human errors in data entry.
2. Sensor or system failures.
3. Data corruption during collection or transfer.
4. Not applicable fields (e.g., “Spouse Name” for unmarried individuals).
5. Intentional non-disclosure (e.g., salary, age, medical details).

-- Why must we handle missing values?

1. Most ML algorithms cannot handle NaN values.
2. Missing data may bias model training if ignored.
3. Dropping too many rows/columns reduces dataset size → model may underfit.
4. Correct handling improves accuracy & generalization.

#### How to Decide Which Strategy to Use?

1. Numeric, symmetric distribution → Mean
2. Numeric, skewed distribution → Median
3. Categorical → Mode
4. Too many missing values (>50%) → Drop column
5. Few missing (<5%) → Safe to impute

## Interpretation:

### Boxplot:

--Points outside the whiskers are outliers.

--The longer whiskers → higher spread of data.

### Histogram:

--Helps check if the column is skewed (common in income data).

--Skewness often causes many outliers on the right side.

```python
from scipy import stats
import numpy as np

# Z-scores for ApplicantIncome
z_scores = stats.zscore(df['LoanAmount'])

# Outliers: less than -3 or greater than +3
outliers = df[(z_scores < -3) | (z_scores > 3)]

print("Number of outliers:", len(outliers))
print(outliers[['LoanAmount']])
```

```python
## OR one line code for z-score
outliers = df[np.abs(z_scores) > 3]
outliers
```

```python
sns.histplot(x=df.ApplicantIncome,kde=True)
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['CoapplicantIncome'].quantile(0.25)
Q3 = df['CoapplicantIncome'].quantile(0.75)

# Interquartile Range
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers_iqr = df[(df['CoapplicantIncome'] < lower_bound) | (df['CoapplicantIncome'] > upper_bound)]

print("Number of outliers:", len(outliers_iqr))
print(outliers_iqr[['CoapplicantIncome']])
```

## Handling Outliers – When to Apply What
### Step 1: Detect Outliers

1. Visualization: Boxplot, Histogram, Scatterplot

2. Statistical Methods:

3. Z-score → if data is normally distributed.

4. IQR (Interquartile Range) → works for skewed distributions.

 Step 2: Handling Outliers (Options & When to Use)
1. Remove Outliers (Trimming)

Simply drop rows with extreme values.

 Use when:

-- Dataset is large (removing a few rows won’t harm).

-- Outliers are due to data entry/measurement errors (e.g., Age = 250).

 Avoid when: dataset is small or outliers are meaningful (e.g., high-income customers).

2. Cap or Floor Outliers (Winsorization)

Replace extreme values with nearest threshold (e.g., 1st & 99th percentile).
Its fine can go with 5% to 10% (0.05 and 0.10)

 Use when:

-- Data is skewed but you don’t want to lose records.

-- Outliers are genuine but too extreme and may distort model performance.

Example: Applicant Income = ₹10,00,000 (replace with 99th percentile value).

3. Transformation

Apply log, square root, or Box-Cox transformation to reduce impact of extreme values.

 Use when:

-- Data is highly skewed (e.g., income, sales).

-- You want to stabilize variance and bring values closer.

 Avoid when: Data already looks symmetric.

4. Imputation with Mean/Median

Replace extreme values with median or mean.

 Use when:

-- Outliers are not many, and data is small.

-- Median works best for skewed data.

Example: Replace income = 10,00,000 with median income = 45,000.

5. Keep Outliers (Do Nothing)

Sometimes outliers are important signals.

 Use when:

Outliers represent rare but valid cases (fraud detection, VIP customers, medical cases).

Business context needs them.

Decision Guide (When to Apply What)
Situation	Recommended Method
Data entry/measurement error	Remove (Trimming)
Small dataset	Winsorization / Median Imputation
Skewed data (e.g., income, sales)	Log / Box-Cox Transformation
Normally distributed data	Z-score detection + Trimming
Business importance in rare values	Keep outliers
Large dataset with extreme distortion	Winsorization or Trimming

 So, the rule is:

Detect → Boxplot, Z-score, IQR.

Decide: Is it an error, rare event, or natural variation?

Apply the method based on dataset size, distribution, and business need.

### SCALING
 -- Scaling is the process of adjusting the range and distribution of numerical features so that they contribute equally in machine learning models.

👉 Without scaling:

Features with larger ranges dominate smaller ones (e.g., income in lakhs vs credit history as 0/1).

Distance-based algorithms get biased.
1. MIN-MAX SCALER
2. STANDARD SCALER
3. ROBUST SCALER

# MinMax Scaling :

- The MinMax Scaler is a popular feature scaling technique in Machine Learning.
- It scales the features of a dataset to a specific range, typically between 0 and 1.
- The main advantage of the MinMax Scaler is that it preserves the shape of the original distribution while bringing the values within a desired range.

- Also known as min-max scaling or min-max normalization, it is the simplest method and consists of rescaling the range of features to scale the range in [0, 1].

```python
- Feature scaling through standardization, also called Z-score normalization, is an important preprocessing step for many machine learning algorithms. It involves rescaling each feature such that it has a standard deviation of 1 and a mean of 0.

- StandardScaler removes the mean and scales each feature/variable to unit variance.
- This operation is performed feature-wise in an independent way.
- StandardScaler can be influenced by outliers (if they exist in the dataset) since it involves the estimation of the empirical mean and standard deviation of each feature.


- Standardization is one of the feature scaling techniques which scales down the data in such a way that the algorithms (like KNN, Logistic Regression, etc.) which are dependent on distance and weights should not get affected by uneven-scaled datasets because if it happens, then, the model accuracy will not be good (will show this practically).
- on the other hand, if we will scale the data evenly in such a way that the data points are mean centric and standard deviation of the distribution is 1 then the weights will be treated equally by the algorithm giving the more relevant and accurate results.

- import numpy as np # linear algebra
- import pandas as pd # data processing
- import matplotlib.pyplot as plt
- import seaborn as sns
- We are importing some required libraries, which will help in practically implementing standardization.

- Numpy: For solving equations related to linear algebra.
- Pandas: For DataFrame manipulation and preprocessing.
- Matplotlib and Seaborn: For data visualization and seeing the distributions.
- df = pd.read_csv('Social_Network_Ads.csv')
- df=df.iloc[:,2:]
- df.sample(5)


### Inference:
- There will always be a debate about whether to do a train test split before doing standard scaling or not, but in my preference, it is essential to do this step before standardization as if we scale the whole data, then for our algorithm, there might be no testing data left which will eventually lead to overfitting condition.
- On the other hand now if we only scale our training data then we will still have the testing data which is unseen for the model.
```

# ROBUST SCALING
- Normalization: shrinks the range such that the range is now between 0 and 1 (or -1 to 1 if there are negative values).
- Robust Scaler: similar to normalization but it instead uses the interquartile range, so that it is robust to outliers.
- It scales features using statistics that are robust to outliers.
- This method removes the median and scales the data in the range between 1st quartile and 3rd quartile. i.e., in between 25th quantile and 75th quantile range.
- This range is also called an Interquartile range(IQR).
- By using RobustScaler(), we can remove the outliers and then use either StandardScaler or MinMaxScaler for preprocessing the dataset.
- It scales features using statistics that are robust to outliers. This method removes the median and scales the data in the range between 1st quartile and 3rd quartile.
- The advantage of using RobustScaler over StandardScaler and MinMaxScaler is that it is more robust to outliers.
- Since it uses median and quartiles instead of mean and variance, it can handle features with a large number of outliers without being influenced by them.
